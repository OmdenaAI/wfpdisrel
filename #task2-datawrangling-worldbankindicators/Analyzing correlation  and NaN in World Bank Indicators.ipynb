## Adapting the dataset to analyze correlation

```python
df_2014 = df[df.columns[[0,2,58]]]
```

```python
df_2014.head()
```

```python
df_2014 = df_2014.pivot(index='Country Name', columns='Series Name', values='2014 [YR2014]')
```

```python
df_2014.index
```

```python
df_2014.corr()
```

```python
c = df_2014.corr() 
s = c.unstack()  # unstacking the correlation matrix 
so = pd.DataFrame(s.sort_values(kind="quicksort", ascending = False)) #sorting and assigning a DaataFrame
```

```python
so.dtypes
```

```python
so.head()
```

```python
#Renaming indexes and after reseting index
so.index.names = ['Series1','Series2']  
so.reset_index()
```

```python
so.head()
```

```python
so.to_csv(r'C:/Users/torre/Documents/Local Omdena WFP files/WorldBank/Part1/WBI_Correlations.csv', index = True)
```
