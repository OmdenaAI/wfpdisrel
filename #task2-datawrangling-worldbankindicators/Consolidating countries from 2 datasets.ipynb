### We are going to check what countries from df1 are found in df2 and viceversa

## Drop countries that are included in ther rows

Inspecting the datasets we see some countries that can be dropped from the Population Pyramid dataset because its information is  inherently in other rows. These are:

- China, Taiwan Province of China  (==> China)
- Western Sahara (==> Morocco)
- State of Palestine (==> +-Israel)

On the other hand, Guadeloupe, Martinique, Mayotte and Réunion are small islands with no information in the WBI dataset (small islands with les than 1,000 population) and will be kept as is.

```python
list_to_drop = ['China, Taiwan Province of China','Western Sahara','State of Palestine']
```

```python
for item in list_to_drop:
    df1 = df1.drop(df1[df1['Country'] == item].index)
```

```python
print ("China, Taiwan Province of China: ",len([x for x in df1['Country'] if 'China, Taiwan Province of China' in x]))
print ("Western Sahara: ",len([x for x in df1['Country'] if 'Western Sahara' in x]))
print ("State of Palestine: ",len([x for x in df1['Country'] if 'State of Palestine' in x]))
```
