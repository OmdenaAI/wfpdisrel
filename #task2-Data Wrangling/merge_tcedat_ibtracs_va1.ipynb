```python
ibtracs_df = pd.read_csv('df_IBTrACS_ISO.csv')
```

```python
ibtracs_df.reset_index(drop=True,inplace=True)
```

```python
ibtracs_df.drop(['Unnamed: 0', 'Unnamed: 0.1'],axis=1, inplace=True)
```

```python
ibtracs_df = ibtracs_df[ibtracs_df['SEASON'] >= 1950]
```

```python
ibtracs_df.reset_index(drop=True, inplace=True)
```

```python
ib_id_set = set(ibtracs_df['SID'].unique())
tc_id_set = set(tcedat_df['IBTrACS_ID'].unique())
```

```python
ibtracs_difference_ids = list(ib_id_set - tc_id_set)
```

```python
for id in ibtracs_difference_ids:
    ibtracs_df = ibtracs_df[ibtracs_df['SID'] != id]
```

```python
ibtracs_df.reset_index(drop=True, inplace=True)
```

```python
ib_id_set = set(ibtracs_df['NAME'].unique())
tc_id_set = set(tcedat_df['TC_name'].unique())
```

```python
len(ib_id_set.intersection(tc_id_set))
```

```python
non_intersection_set = (ib_id_set | tc_id_set) - (ib_id_set & tc_id_set)
len(non_intersection_set)
```

```python
for element in list(non_intersection_set):
    ibtracs_df = ibtracs_df[ibtracs_df['NAME'] != element]
    tcedat_df = tcedat_df[tcedat_df['TC_name'] != element]
```

```python
ibtracs_df.reset_index(drop=True, inplace=True)
tcedat_df.reset_index(drop=True, inplace=True)
```

```python
merged_df = tcedat_df.merge(ibtracs_df, left_on=['year', 'IBTrACS_ID','TC_name', 'ISO3'], right_on=['SEASON', 'SID', 'NAME','ISO'], how='inner')
```
