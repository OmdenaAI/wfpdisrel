## Selection of events from 1980+

```python
df = dft[dft['year']>= 1980]
df.reset_index(drop=True,inplace=True)
df
```

```python
MAX_WIND = df[['WMO_WIND','USA_WIND','TOK_WIND','CMA_WIND',
               'HKO_WIND','NEW_WIND','REU_WIND','BOM_WIND',
               'NAD_WIND','WEL_WIND','DS8_WIND','TD6_WIND',
               'TD5_WIND','NEU_WIND','MLC_WIND']].max(axis=1,skipna=True)
df['MAX_WIND'] = MAX_WIND
df
```

### Drop records where 'MAX_WIND' = NaN

```python
df= df[df['MAX_WIND'].notna()]
```

## Selection by minimum distance to land

```python
years = df['year'].unique()

idx_min_list = []

for year in years:
    aux1_df = df[df['year']==year]
    aux1_number = aux1_df['NUMBER'].unique()
    for number in aux1_number:
        aux2_df = aux1_df[aux1_df['NUMBER']==number]
        #index of first occurrence of minimum over requested axis
        idx_min = aux2_df['DIST2LAND'].idxmin()
        idx_min_list.append(idx_min)
```

```python
df_2land = df.loc[idx_min_list]
df_2land.reset_index(drop=True,inplace=True)
df_2land
```

```python
sel_columns = ['year','ISO_TIME',
               'BASIN','SUBBASIN','LAT','LON','MAX_WIND',
               'USA_SSHS',
               'TOK_GRADE','CMA_CAT','HKO_CAT','NEW_GRADE',
               'BOM_TYPE','NAD_CAT','TD6_STAGE','NEU_CLASS','MLC_CLASS',               
               'NAME']
df_2land = df_2land[sel_columns]
df_2land
```

## Selection by maximum wind speed

```python
years = df['year'].unique()

idx_max_list = []

for year in years:
    aux1_df = df[df['year']==year]
    aux1_number = aux1_df['NUMBER'].unique()
    for number in aux1_number:
        aux2_df = aux1_df[aux1_df['NUMBER']==number]
        #index of first occurrence of maximum over requested axis
        idx_max = aux2_df['MAX_WIND'].idxmax()
        idx_max_list.append(idx_max)
```

```python
df_maxw = df.loc[idx_max_list]
df_maxw.reset_index(drop=True,inplace=True)
df_maxw = df_maxw[sel_columns]
df_maxw
```
