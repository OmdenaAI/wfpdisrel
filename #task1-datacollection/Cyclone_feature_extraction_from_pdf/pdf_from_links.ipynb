# This notebook extracts information from PDFs extracted via the index file. 

## Step 1: Extract and download 'PDFs' only. Documents prior to 1995 are images so I hae ignored those.   

### The index XML file is located: https://www.nhc.noaa.gov/TCR_StormReportsIndex.xml

### Use wget, xml etree and pandas to download all the PDF files

```python
import wget
import pandas as pd 
import xml.etree.ElementTree as et 
import requests

from pathlib import Path
```

```python
xtree = et.parse("TCR_StormPDF_1995.xml")
xroot = xtree.getroot() 

df_cols = ["StormName", "StormReportURL", "Year", "Basin"]
rows = []

for node in xroot: 
    StormName = node.find("StormName").text if node is not None else None
    StormReportURL = node.find("StormReportURL").text if node is not None else None
    Year = node.find("Year").text if node is not None else None
    Basin = node.find("Basin").text if node is not None else None
    
    if(StormReportURL.endswith('.pdf')):    
        rows.append({"StormName": StormName,
                     "StormReportURL": StormReportURL, 
                     "Year": Year, 
                     "Basin": Basin})

out_df = pd.DataFrame(rows, columns = df_cols)
```

```python
out_df.to_csv('All data1.csv')
```

```python
url = 'https://www.nhc.noaa.gov/data/tcr/AL062018_Florence.pdf'
filepath = '1995_pdf/' + url.split('/')[-1]
filename = Path(filepath)    
response = requests.get(url)
filename.write_bytes(response.content)
```

```python
for index, storm_record in out_df.iterrows():
    if(index>=201):
        url = storm_record['StormReportURL']
        fname = url.split('/')[-1]
        filepath = '1995_pdf/' + fname
        filename = Path(filepath)    
        response = requests.get(url)
        filename.write_bytes(response.content)
        print(f'Download number {index}: File: {fname} completed.')
```
