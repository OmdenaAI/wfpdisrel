```python
#Renamig TCE-DAT columns
df1.rename(columns={'ISO3':'Country Code'}, inplace = True)
df1.rename(columns={'year':'Year'}, inplace = True)
```

```python
#Dropping some columns of total population dataset as it is required to change it into melted form.
df5.drop("Indicator Name", axis = 1,inplace = True)
df5.drop("Indicator Code", axis = 1,inplace = True)
df5.drop("Country Name", axis = 1,inplace = True)
```

```python
tot_pop = pd.melt(df5,id_vars = ["Country Code"], var_name = "Year" , value_name ="Tot_pop")
# as in TCE-DAT we have data till 2015 so taking total population data till 2015 
tot_pop=tot_pop[tot_pop['Year']<='2015']
df1['Year'] = df1['Year'].astype('int')
df1['Country Code'] = df1['Country Code'].astype('str')
tot_pop['Year'] = tot_pop['Year'].astype('int')
tot_pop['Country Code'] = tot_pop['Country Code'].astype('str')
# As length of tropical cyclone data and melted data of total population is not equal so using slicing 
# to merge them on the basis of two keys (Country Code and Year)
df6 = tot_pop.iloc[0:5358, :]
df_merged1 = pd.merge(df6,df1,how = 'inner',left_on = ['Country Code','Year'],right_on = ['Country Code','Year'])
df7 = tot_pop.iloc[5358:10716, :]
df_merged2 = pd.merge(df7,df1,how = 'inner',left_on = ['Country Code','Year'],right_on = ['Country Code','Year'])
df8 = tot_pop.iloc[9426:, :]
df_merged3 = pd.merge(df8,df1,how = 'inner',left_on = ['Country Code','Year'],right_on = ['Country Code','Year'])
#tropical cyclone data merging it with total population data
bigdata1 = pd.concat([df_merged1,df_merged2,df_merged3], ignore_index=True, sort =False)
```

```python
# Dropping duplicates
bigdata1 = bigdata1.drop_duplicates()
bigdata1
```
