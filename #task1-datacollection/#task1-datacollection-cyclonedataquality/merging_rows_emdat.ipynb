## To have sum 

```python
columns_to_group_by = ['Year', 'ISO', 'Location', 'Start year', 'Start month',
                   'End year', 'End month'] # without 'End day', 'Start day'

columns_to_count = ['Total deaths', 'Num injured',
                   'Num affected', 'Num homeless', 'Total affected',
                   "Insured losses (\'000$)", "Insured losses (\'000$)", "Total damage (\'000$)"]

columns_to_leave = columns_to_group_by + columns_to_count
columns_to_leave
```

```python
my_df = df[columns_to_leave]
d1 = my_df[ my_df["ISO"] == "AUS"]
d1[ d1["Year"] == 1980]
```

```python
my_df.ffill()
new_df = df.groupby(columns_to_group_by, as_index=False).sum()
```

```python
d1 = new_df[ new_df["ISO"] == "AUS"]
d2 = d1[ d1["Year"]== 1980]
d2
```

## To calculate magnitude

```python
columns_to_count2 = ["Magnitude (value)"]

columns_to_leave2 = columns_to_group_by + columns_to_count2
columns_to_leave2
```

```python
my_df2 = df[columns_to_leave2]
d1 = my_df2[ my_df2["ISO"] == "AUS"]
d1[ d1["Year"] == 1980]
```

```python
my_df2.ffill()
new_df_mean = my_df2.groupby(columns_to_group_by, as_index=False).mean()
new_df_mean.rename(columns={"Magnitude (value)": "Magnitude (mean value)"}, inplace=True)
```

```python
d1 = new_df_mean[ new_df_mean["ISO"] == "AUS"]
d2 = d1[ d1["Year"]== 1980]
d2
```

```python
new_df_max = my_df2.groupby(columns_to_group_by, as_index=False).max()
new_df_max.rename(columns={"Magnitude (value)": "Magnitude (max value)"}, inplace=True)
d1 = new_df_max[ new_df_max["ISO"] == "AUS"]
d2 = d1[ d1["Year"]== 1980]
d2
```

```python
new_df_min = my_df2.groupby(columns_to_group_by, as_index=False).min()
new_df_min.rename(columns={"Magnitude (value)": "Magnitude (min value)"}, inplace=True)
d1 = new_df_min[ new_df_min["ISO"] == "AUS"]
d2 = d1[ d1["Year"]== 1980]
d2
```

## Merging datasets

```python
new_df_min_max = new_df_min.merge(new_df_max, on=columns_to_group_by, how='inner')
new_df_min_max.shape
```

```python
new_df_magnitude = new_df_min_max.merge(new_df_mean, on=columns_to_group_by, how='inner')
new_df_magnitude.shape
```

```python
final_df = new_df_magnitude.merge(new_df, on=columns_to_group_by, how='inner')
final_df.shape
```
